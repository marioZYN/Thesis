```python
def get_data(tid, path, r):

    coords_df = pd.read_csv(path+'coords.csv')
    img = cv2.cvtColor(cv2.imread(path + 'Train/' + str(tid) + '.jpg'), cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (int(img.shape[1] * r), int(img.shape[0] * r)))

    rows = (coords_df[coords_df['tid'] == tid].row * r).astype(int)
    cols = (coords_df[coords_df['tid'] == tid].col * r).astype(int)
    cls = coords_df[coords_df['tid'] == tid].cls

    y = np.zeros((*img.shape[:-1], 5))
    y[rows, cols, cls] = 1
    
    img = np.pad(img, ((31, 31), (31, 31), (0, 0)), mode='constant', constant_values=0)
    y = np.pad(y, ((31, 31), (31, 31), (0, 0)), mode='constant', constant_values=0)

    return img, y
```

```python
# define model
x = Input(shape=(None, None, 5))
t = Conv2D(5, (32, 32), padding='valid')(x)
model = Model(inputs=x, outputs=t)

filters = np.zeros((32, 32, 5, 5))
for i in range(5):
    f = np.zeros((32, 32, 5))
    f[:, :, i] = 1
    filters[:, :, :, i] = f
model.layers[-1].set_weights([filters, np.zeros(5)])

model.layers[-1].trainable = False
```

```python
# get data for testing
idx = 0
r = 0.2
train_df = pd.read_csv('/Users/YINAN/Local/Sealions/Data/train.csv', index_col='train_id')

img, y = get_data(idx, '/Users/YINAN/Local/Sealions/Data/', r)
print('input shape: img {}, y {}'.format(img.shape, y.shape))

plt.figure(figsize=(12, 12))
plt.subplot(1, 1, 1)
plt.imshow(img)
plt.title('original image with padding')

plt.figure(figsize=(12, 6))
print('\nreal sea lion counts:')
print(train_df.loc[idx])
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(y[:,:,i])
    plt.title(setting.class_names[i] + ": " + str(int(y[:, :, i].sum())))

res = model.predict(y[np.newaxis, :])
plt.figure(figsize=(12, 6))
for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(res.squeeze()[:,:,i])
    v = int(res.squeeze()[:, :, i].sum())
    plt.title('{} heatmp : {} / {}'.format(setting.class_names[i], v, v//(32**2)))
```
